## Step 1: Import Libraries
We begin by importing the necessary libraries for data analysis and visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a consistent style for visualizations
sns.set_style("whitegrid")
print("Libraries loaded successfully.")
```

## Step 2: Load the Dataset
dataset from the CSV file. The encoding is set to 'ISO-8859-1' to handle special characters in the data

```python
df = pd.read_csv("Most Streamed Spotify Songs 2024.csv", encoding='ISO-8859-1')
print("Dataset loaded successfully.")
df.head()
```

## Step 3.2: Data Overview
Analyzing the data structure to better understand the database

```python
# Preview the first few rows
df.head()
```

```python
# Overview of data types and structure
df.info()
```

### Insights from Data Overview
- The dataset contains 4,600 songs with 29 columns, including track name, artist, release date, Spotify streams, YouTube views, and more.
- Some columns have missing values (e.g., `TIDAL Popularity` is completely empty, and `Soundcloud Streams` has only 1,267 non-null entries).
- Most columns are of type `object`, indicating that some data (like streams and views) may need to be cleaned and converted to numeric types for analysis

## Step 3.3: Clean and transform data
I will clean the data to make it suitable for analysis by standardizing column names, converting relevant columns to numeric types, and handling missing or duplicate data

```python
# Standardize column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# Define cleaning function for numeric fields
def clean_to_numeric(series):
    return (series.astype(str)
            .str.replace(r'[^\d.]', '', regex=True)
            .replace('', None)
            .astype(float))

# Clean key columns
df['spotify_streams'] = clean_to_numeric(df['spotify_streams'])
df['youtube_views'] = clean_to_numeric(df['youtube_views'])

# Drop missing or duplicated rows
df.dropna(subset=['spotify_streams', 'youtube_views'], inplace=True)
df.drop_duplicates(inplace=True)

# Verify cleaning
df[['spotify_streams', 'youtube_views']].info()
```

```python

print("\nDescriptive Statistics (Before Cleaning):\n")
print(df.describe(include='all'))


print("\nMissing Values (Before Cleaning):\n")
print(df.isnull().sum())
```

```python

if 'release_date' in df.columns:
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
    print("\\n'release_date' column converted to datetime.")
    print(f"Number of NaT values created: {df['release_date'].isnull().sum()}")
else:
    print("Warning: 'release_date' column not found.")
```

### Insights from Data Cleaning
- After cleaning, the dataset is reduced to 4,216 songs (from 4,600) due to missing values and duplicates
- The `spotify_streams` and `youtube_views` columns are now in a numeric format (float64) with no missing values, making them ready for analysis

## Step 3.4: Descriptive Statistics and Correlation
I will perform descriptive statistics and correlations to understand the distribution and relationships in the data

```python
# Summary statistics
df[['spotify_streams', 'youtube_views']].describe()
```

### Insights from Descriptive Statistics
- **Spotify Streams**: The average number of streams is 443 million, with a high standard deviation (525 million), indicating significant variation. The maximum stream per song is 4.28 billion.
- **YouTube Views**: The average number of views is 395 million, with a maximum of 16.32 billion, indicating that some songs achieve much higher views on YouTube.
- The data is highly skewed, as the median (50th percentile) for both columns is much lower than the average.

```python
# Correlation
df[['spotify_streams', 'youtube_views']].corr()
```

Insights from Correlation
There is a moderate positive correlation (~0.47) between Spotify streams and YouTube views.
Suggesting that songs with high streams on Spotify also tend to have higher views on YouTube, but the relationship is not particularly strong, suggesting that other factors could be influencing popularity across the platform.

## Step 3.5: Cumulative Analysis
Analyze the trend of Spotify streams over the years to identify patterns in song popularity based on release year

```python
# Total and average streams by artist
artist_summary = df.groupby('artist').agg({
    'spotify_streams': ['sum', 'mean'],
    'youtube_views': ['sum', 'mean']
}).sort_values(('spotify_streams', 'sum'), ascending=False).head(10)
artist_summary
```

This line chart shows the total number of Spotify streams per release year. It helps identify trends in listener behavior over time

**Insight:** There is a noticeable increase in total streams in recent years, indicating growth in the platformâ€™s user base or an increasing preference for newer music


## Step 3.6: Data Visualizations

```python
# Pie Chart of Top 5 Artists
top_artists = df.groupby('artist')['spotify_streams'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,8))
plt.pie(top_artists, labels=top_artists.index, autopct='%1.1f%%', startangle=140)
plt.title('Share of Spotify Streams by Top 5 Artists')
plt.axis('equal')
plt.show()
```

Who controls Spotify?
The Pie Chart shows that 5 artists control a large share of the airplay, with ~20% each.

**Insight:**  Music on the platform is very concentrated in the hands of top artists. Itâ€™s a market with big, dominant stars

```python
# Top 10 Most Streamed Songs
top_songs = df.sort_values(by='spotify_streams', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_songs, y='track', x='spotify_streams', palette='viridis')
plt.title('Top 10 Most Streamed Songs on Spotify (2024)')
plt.xlabel('Spotify Streams')
plt.ylabel('Track')
plt.tight_layout()
plt.show()
```

This bar chart displays the top 10 most streamed songs on Spotify.

**Insight:** A small group of songs stands out with extremely high stream counts, highlighting their exceptional popularity


```python
# Histogram of Spotify Streams
plt.figure(figsize=(10,6))
sns.histplot(df['spotify_streams'], bins=50, kde=True, color='orange')
plt.title('Distribution of Spotify Streams')
plt.xlabel('Spotify Streams')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

This histogram shows the distribution of Spotify streams across all songs.

**Insight:** Most songs have relatively low stream counts, while only a few achieve extremely high numbers â€” a typical "long tail" pattern in digital media.



```python
# Scatterplot: Spotify vs YouTube
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='youtube_views', y='spotify_streams', alpha=0.5)
plt.title('Spotify Streams vs YouTube Views')
plt.xlabel('YouTube Views')
plt.ylabel('Spotify Streams')
plt.tight_layout()
plt.show()
```

YouTube â‰  Spotify
The scatterplot (Spotify vs YouTube) shows that there is no strong correlation between the number of views on YouTube and plays on Spotify.

**Insight:** Success on YouTube does not guarantee success on Spotify (and vice versa). These are different audiences/behavior

```python
# Line chart of Spotify streams over years
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_year'] = df['release_date'].dt.year
yearly = df.groupby('release_year')['spotify_streams'].sum().reset_index()

plt.figure(figsize=(10,6))
plt.plot(yearly['release_year'], yearly['spotify_streams'], color='orange', linewidth=2)
plt.scatter(yearly['release_year'], yearly['spotify_streams'], color='blue', s=50)
plt.title('Spotify Streams Over Time')
plt.xlabel('Release Year')
plt.ylabel('Total Streams')
plt.tight_layout()
plt.show()
```

Past songs â‰  streaming hits
The time graph shows that since 2015 there has been a sharp increase in the amount of plays, with peaks around 2020.

**Insight:**  Streaming is a modern phenomenon â€“ new songs receive much more plays
